STATISTICAL ANALYSIS

Now we will find if there is any correlation between the mass of white dwarfs and the metals that are visible on their surface. This can be done using the Pearson correlation. This gives 2 outputs, namely correlation and p-value. Correlation values lie from -1 to +1. 
1. If correlation = 0 : No correlation between the quantities. They are independent.
2. If correlation > 0 : They have a positive correlation, i.e, if one of the values increases then the other one also increases. The nearer the value is to +1, the stronger their relation is.
2. If correlation <> 0 : They have a negative correlation, i.e, if one of the values increases then the other one decreases. The nearer the value is to -1, the stronger their relation is.



P-value tells whether or not the result is statistically significant. Generally, if p-value is very small then the result is statistically significant. 
There are 2 types of hypothesis: Null and Alternative. 
1. Null hypothesis        (H₀)= It assumes that there exists no relationship between the two variables.
2. Alternative hypothesis (H₁)= It assumes that null hypothesis is false, i.e, there is a relation between the two variables.

Small p-value (typically ≤ 0.05): Indicates strong evidence against the null hypothesis, leading to its rejection. This proves that there exists a relationship between variables.

Large p-value (typically > 0.05): Indicates weak evidence against the null hypothesis, leading to its acceptance (not rejection). This might mean that the variables are independent.

Therefore we say that a result is significant if the correlation value is non zero and preferrably closer to +1 or -1 (depending on the situation) AND the p-value is ≤ 0.05. 

We will find pearson correlation between "Mass values" and the "log ratios of metals with Helium or Hydrogen"



PEARSON CORRELATION BETWEEN "MASS" AND "METAL ABUNBDANCE"

```python
for i in range(len(clean_data_functions_dict)):
    print(f"""
          FOR THE {clean_data_names_dict.get(i)}
                                                """)
    for index, column in enumerate(clean_data_functions_dict.get(i).columns[first_and_last_column_indices[i][0]:first_and_last_column_indices[i][1] + 1]):       #this for loop gives index number to the columns from the starting column to the ending column in each file
        dz_clean_loc = clean_data_functions_dict.get(i).loc[:,column]
    
        def columns_correlation_with_mass(column) :
            drop_NaN_values =clean_data_functions_dict.get(i)[["mass",column]].dropna(axis=0)     # Select the 'mass' and the current column from the dataset. Drop the NaN values if present in any of their rows
            drop_NaN_values["mass"] = pd.to_numeric(drop_NaN_values["mass"], errors='coerce')     # Convert both columns to numeric values,so non-numeric values like "<4.876" etc get converted into NaN
            drop_NaN_values[column] = pd.to_numeric(drop_NaN_values[column], errors='coerce')

            drop_NaN_values_again = drop_NaN_values.dropna(axis=0)                                # Drop rows with newly formed NaN values after conversion

            return drop_NaN_values_again
        
        if len(columns_correlation_with_mass(column=column)) < 2:                               # To not consider the faulty data
            print(f"Not enough data for column {column}")
            continue

        correlation_mass, p_value_mass = pearsonr(columns_correlation_with_mass(column)["mass"], columns_correlation_with_mass(column)[column])
        print(f"For {column} and mass parameter :Pearson correlation: {correlation_mass}, P-value: {p_value_mass}")

        correlation_matrix = columns_correlation_with_mass(column)[["mass",column]].corr(numeric_only=True)    # For plotting correlation values
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
        plt.title(f'Correlation between Mass and Metal Pollution for {column}')
        plt.show()
```

So, analysing both the correlation and p-values, we observe that only a few metal ratios have a significant correlation with mass of white dwarfs and are also statistically significan. These metals are :
1. For DZ : Ca/He and Mg/He
2. For DAZ : Ni/H
3. For DBZ : Ca/He

We will call these metal values as "SIGNIFICANT METAL" values here onwards.
