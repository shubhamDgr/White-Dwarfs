```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from data_clean import dz_clean, daz_clean, dbz_clean
from scipy.stats import pearsonr
import numpy as np
```

```python
clean_data_functions_dict = {0 :dz_clean(),
                             1 :daz_clean(),
                             2 :dbz_clean()}
clean_data_names_dict = {0 :"DZ White Dwarfs",
                         1 :"DAZ White Dwarfs",
                         2 :"DBZ White Dwarfs"}
significant_metals_for_each_type = {0 :["logcahe","logmghe"],                   # These are the only metals which show significant correlation with mass as per the results obtained above. So we will analyze them further
                                    1 :"lognih",
                                    2 :"logcahe"}
```

```python
parameter_list = ["teff","logg","mass","age"] 
```

VISUALIZATION OF IMPACT OF EACH PARAMETER


Now we will plot the metal abundance on Y-axis and all the 4 parameter in the "parameter_list" on X-axis, one by one. This way we can visualize how each parameter affects the metal abundance individually for different types of white dwarfs.

```python
for i in range(len(clean_data_functions_dict)):
    if isinstance(significant_metals_for_each_type.get(i),list):                                                # when i= 0, significant_metals_for_each_type.get(i) returns the list ["logcahe","logmghe"]. This code handles i=0 case
        for j in parameter_list:                                                                                # plots graph with each parameter in parameter_list one by one
            sns.scatterplot(data=clean_data_functions_dict.get(i), x=j, y=f"{significant_metals_for_each_type.get(i)[0]}", hue=j)
            plt.title(f"Metal abundance vs {j} in {clean_data_names_dict[i]}")
            plt.show()
            
            sns.scatterplot(data=clean_data_functions_dict.get(i), x=j, y=f"{significant_metals_for_each_type.get(i)[1]}", hue=j)
            plt.title(f"Metal abundance vs {j} in {clean_data_names_dict[i]}")
            plt.show()

    elif isinstance(significant_metals_for_each_type.get(i),str):                                               # when i=1 or i=2, significant_metals_for_each_type.get(i)  returns a string. so this elif handles these 2 cases                           
        for j in parameter_list:
            sns.scatterplot(data=clean_data_functions_dict.get(i), x=j, y=f"{significant_metals_for_each_type.get(i)}", hue=j) 
            plt.title(f"Metal abundance vs {j} in {clean_data_names_dict[i]}")
            plt.show()
```
